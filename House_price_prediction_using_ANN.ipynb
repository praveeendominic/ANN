```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
```

```python
data= fetch_california_housing()
```

```python
X=pd.DataFrame(data.data, columns=data.feature_names)
X.head()

X=X.iloc[:,:-2]
```

```python
y=pd.DataFrame(data.target, columns=data.target_names)
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

X_train_scaled=sc.fit_transform(X_train)
X_test_scaled=sc.transform(X_test)
```

```python
X_train_scaled.shape
```

```python
y.MedHouseVal.values
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense
from keras.models import Sequential
```

```python
LAYERS = [Dense(20,input_shape=(6,), activation=tf.keras.activations.relu, name='hidden_layer1'),
          Dense(20, activation = tf.keras.activations.relu, name='hidden_layer2'),
          Dense(20, activation = tf.keras.activations.relu, name='hidden_layer3'),
          Dense(1, name='op_layer')]

OPTIMIZER = tf.keras.optimizers.Adam(1e-3)
LOSS = tf.keras.losses.MSE
```

```python
model = Sequential(LAYERS)
model.summary()
```

```python
model.layers
```

```python
for layer in model.layers:
  print(layer.variables)
```

```python
weights,bias=model.layers[0].get_weights()
```

```python
model.compile(optimizer=OPTIMIZER, loss='mse', metrics=['accuracy'] )
```

```python
model.fit(X_train_scaled, y_train, epochs=50, batch_size=32, verbose=2, validation_split=0.10)
```
